## Part 1 : Dataset Creation from Mathlib Lean4
----------------

### Dataset Creation

```python
import json

file_paths = []

testf = open('data\\leandojo_benchmark_4\\random\\test.json')
trainf = open('data\\leandojo_benchmark_4\\random\\train.json')

test_data = json.load(testf)
train_data = json.load(trainf)

for dat in test_data:
    file_paths.append(dat['file_path'])

for dat in train_data:
    file_paths.append(dat['file_path'])

testf.close()
trainf.close()

file_paths = set(file_paths)
file_paths = sorted(file_paths)
```

```python
theorems = []
splits = ["\n\n", "\n  |"]

num = 0
bad = 0

for file in file_paths:
    num += 1

    try:
        f = open(file, "r", encoding="utf8")
    except:
        print("File " + str(num) + ": " + file)
        bad += 1
        continue

    
    fc = f.read()

    start = fc.find("/-")
    end = start + fc[start:].find("-/")

    while start > -1:
        # print("comment " + str(len(fc)) + ": " + str(start) + " " + str(end))
        fc = fc[0:start] + fc[end+2:]

        start = fc.find("/-")
        end = start + fc[start:].find("-/")
    
    start = fc.find("--")
    end = start + fc[start:].find("\n")

    while start > -1:
        # print("comment " + str(len(fc)) + ": " + str(start) + " " + str(end))
        fc = fc[0:start] + fc[end+1:]

        start = fc.find("--")
        end = start + fc[start:].find("\n")
        
    temp = fc.split("theorem")
    temp = temp[1:]

    for temp_thm in temp:
        temp_thm = "theorem" + temp_thm
        
        for split in splits:
            temp_thm = temp_thm[:temp_thm.find(split)]
        
        end = temp_thm.find(" :=")

        if (end == -1):
            theorems.append((file, temp_thm))
        else:
            theorems.append((file, temp_thm[:end]))
    
    f.close()
```

```python
print("Total files: " + str(num))
print("Bad files: " + str(bad))
print("\nTotal number of theorems found: " + str(len(theorems)))
```

```python
theorems = sorted(theorems, key=lambda x: len(x[1]))
theorems.reverse()

for i in range(1000):
    file, theorem = theorems[i]
    print(file + ": " + str(len(theorem)))
```

```python
for i in range(len(theorems)):
    name, thm = theorems[i]

    name = name.split("/")[1:]
    name[-1] = name[-1][:name[-1].find(".")]
    
    theorems[i] = (name, thm)
```

```python
labels = []
thms = []

for (label, thm) in theorems:
    labels.append(label[0])
    thms.append(thm)

unique_values_set = set(labels)
unique_values_list = list(unique_values_set) 

numHits = [0 for _ in range(len(labels))]

for i in range (len(labels)):
    for j in range(len(unique_values_list)):
        if (labels[i] == unique_values_list[j]):
            numHits[j] += 1

label_sizes = []

for i in range (len(unique_values_list)):
    print(unique_values_list[i], numHits[i])
    label_sizes.append((unique_values_list[i], numHits[i]))
```

```python
num_limited_classes = 10

label_sizes = sorted(label_sizes, key=lambda x: x[1])
label_sizes.reverse()
limited_labels = []

rank = 1

for (label, hits) in label_sizes[:num_limited_classes]:
    limited_labels.append(label)
    print(str(rank) + ".", label, hits)

    rank += 1
```

```python
limited_theorems = []

for (label, thm) in theorems:
    if label[0] in limited_labels:
        limited_theorems.append((label, thm))

print("\nTotal number of classes: " + str(len(unique_values_list)))
print("Total number of limited classes: " + str(num_limited_classes))

print("\nTotal number of theorems: " + str(len(theorems)))
print("Total number of theorems in top " + str(num_limited_classes) + " largest classes: " + str(len(limited_theorems)))
```

### CSV File Creation

```python
maxLen = 0

for (label, thm) in theorems:
    if (len(label) > maxLen):
        maxLen = len(label)

field = []

for i in range(1, maxLen+1):
    field.append("subject_label" + str(i))

field.append("theorem")
```

```python
import csv

csv_type = 2

with open('all_theorem_classes_type' + str(csv_type) + '.csv', 'w', encoding="utf8", newline='') as file:
    writer = csv.writer(file)    

    if (csv_type == 1):
        writer.writerow(["subject_label", "theorem"])

        for (label, thm) in theorems:
            writer.writerow([label[0], thm])
    elif (csv_type == 2):
        writer.writerow(field)

        for (label, thm) in theorems:
            while (len(label) < maxLen):
                label.append("")
            
            writer.writerow(label + [thm])

with open('limited_theorem_classes_type' + str(csv_type) + '.csv', 'w', encoding="utf8", newline='') as file:
    writer = csv.writer(file)

    if (csv_type == 1):
        writer.writerow(["subject_label", "theorem"])

        for (label, thm) in limited_theorems:
            writer.writerow([label[0], thm])
    elif (csv_type == 2):
        writer.writerow(field)

        for (label, thm) in limited_theorems:
            while (len(label) < maxLen):
                label.append("")
            
            writer.writerow(label + [thm])
```
